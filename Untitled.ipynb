```python
import sympy as sp
import scipy.integrate
```

```python
x = sp.Function('x', real=True)
k, m, t, A, phi, omega = sp.symbols('k m t A phi omega')
```

```python
eq = sp.Eq(m*sp.diff(x(t),t,t), -k*x(t))
eq
```

```python
sol = sp.Eq(x(t), A*sp.sin(omega*t+phi))
sol
```

```python
sp.plot(sol.rhs.subs({A:1,omega:1, phi:0}), sol.rhs.subs({A:1,omega:2, phi:0}))
```

```python
y = sp.symbols("y0 y1", cls=sp.Function)
```

```python
eq.subs({sp.diff(x(t),t):y[0](t), x(t):y[1](t)})
```

```python
sys = [sp.Eq(sp.diff(y[0](t),t), sp.solve(eq.subs({sp.diff(x(t),t):y[0](t), x(t):y[1](t)}), sp.diff(y[0](t),t))[0]), sp.Eq(sp.diff(y[1](t),t), y[0](t))]
sys
```

```python
f = sp.lambdify((k, m, y[1], t),sys[0].rhs)
```

```python
scipy.integrate.solve_ivp(f, t_span=)
```
